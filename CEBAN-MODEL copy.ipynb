```python
def clean_age(age_str):
    try:
        # If it's a single number (e.g., '23')
        return float(age_str)
    except ValueError:
        # If it's a range (e.g., '21 - 22', '18 -20')
        # Remove spaces around the hyphen for consistent splitting
        parts = age_str.replace(' ', '').split('-')
        if len(parts) == 2:
            try:
                lower = float(parts[0])
                upper = float(parts[1])
                return (lower + upper) / 2
            except ValueError:
                return np.nan # Return NaN if parts are not valid numbers
        else:
            return np.nan # Return NaN for other unexpected formats

# Apply the cleaning function to the 'age' column
df_new_column['age_cleaned'] = df_new_column['age'].apply(clean_age)

print("\n'age' column after cleaning:")
print(df_new_column[['age', 'age_cleaned']])
print("\nData type of 'age_cleaned':", df_new_column['age_cleaned'].dtype)
```

```python
def clean_dependents(dep_str):
    # Standardize 'Tidak' to '0' or a specific numeric representation
    if dep_str.lower() == 'tidak':
        return 0  # Assuming 'Tidak' means 0 dependents

    try:
        # If it's a single number
        return float(dep_str)
    except ValueError:
        # If it's a range (e.g., '1-2', '3-4', '5-6')
        parts = dep_str.replace(' ', '').split('-')
        if len(parts) == 2:
            try:
                lower = float(parts[0])
                upper = float(parts[1])
                return (lower + upper) / 2 # Taking the mean for the range
            except ValueError:
                return np.nan # Return NaN for unparseable parts
        else:
            return np.nan # Return NaN for other unexpected formats

# Apply the cleaning function to the 'dependents' column
df_new_column['dependents_cleaned'] = df_new_column['dependents'].apply(clean_dependents)

print("\n'dependents' column after cleaning:")
print(df_new_column[['dependents', 'dependents_cleaned']])
print("\nData type of 'dependents_cleaned':", df_new_column['dependents_cleaned'].dtype)
```

```python
def clean_household_size(household_str):
    """
    Cleans the 'household_size' string, converting it to a float.
    Handles single numbers, ranges (takes the mean), and 'Tidak' (converts to 0).
    Returns np.nan for unparseable formats.
    """
    # Standardize 'Tidak' to 0 (assuming 'Tidak' means 0 household members, or 'not applicable'
    # which we map to 0 for numerical conversion)
    if str(household_str).lower() == 'tidak':
        return 0.0 # Return as float for consistency

    try:
        # If it's a single number (e.g., '5', '7')
        return float(household_str)
    except ValueError:
        # If it's a range (e.g., '1-2', '5-6')
        # Remove spaces around the hyphen for consistent splitting
        parts = str(household_str).replace(' ', '').split('-')
        if len(parts) == 2:
            try:
                lower = float(parts[0])
                upper = float(parts[1])
                return (lower + upper) / 2 # Take the mean of the range
            except ValueError:
                return np.nan # Return NaN if parts are not valid numbers
        else:
            return np.nan # Return NaN for other unexpected formats (e.g., malformed strings)

print("Original household_size column:")
print(df_new_column['household_size'])
print(f"Original dtype: {df_new_column['household_size'].dtype}")
print(f"Original unique values: {df_new_column['household_size'].unique()}")

# Apply the new cleaning function
df_new_column['cleaned_household_size'] = df_new_column['household_size'].apply(clean_household_size)
```

```python
# --- 3. Fungsi Preprocessing Utama untuk Satu Titik Data Baru ---
def preprocess_single_new_data(raw_data_dict):
    """
    Menerima satu set data mentah sebagai dictionary (merepresentasikan satu individu),
    mengaplikasikan semua langkah preprocessing yang identik dengan yang digunakan saat training,
    dan mengembalikan DataFrame yang siap untuk dimasukkan ke model.

    Argumen:
        raw_data_dict (dict): Sebuah dictionary yang berisi data mentah satu individu.
                              Kunci harus sesuai dengan nama kolom asli sebelum preprocessing.
    Returns:
        pd.DataFrame: DataFrame 1 baris dengan fitur-fitur yang sudah diproses,
                      siap untuk dimasukkan ke model.
    """
    # Konversi dictionary ke DataFrame 1 baris
    processed_df = pd.DataFrame([raw_data_dict])

    # --- Terapkan pembersihan awal untuk 'age', 'dependents', 'household_size' ---
    processed_df['age_cleaned'] = processed_df['age'].apply(clean_age)
    processed_df['dependents_cleaned'] = processed_df['dependents'].apply(clean_dependents)
    processed_df['household_size_cleaned'] = processed_df['household_size'].apply(clean_household_size)
    processed_df = processed_df.drop(columns=['age', 'dependents', 'household_size'], errors='ignore')

    # Tangani NaNs di kolom yang sudah dibersihkan (penting: gunakan mean/median dari data training jika disimpan)
    # Untuk POC, jika tidak disimpan, Anda bisa menggunakan 0 atau median/mean dari kolom ini sendiri.
    # Namun, untuk produksi, nilai imputasi harus dari data training.
    for col in ['age_cleaned', 'dependents_cleaned', 'household_size_cleaned']:
        if col in processed_df.columns:
            processed_df[col].fillna(0, inplace=True) # Ganti 0 dengan mean/median dari training jika tersedia

    # education_level TIDAK ADA, jadi tidak ada Ordinal Encoding di sini

    # --- One-Hot Encoding untuk fitur kategorikal lainnya (misal: 'gender') ---
    # Identifikasi kolom 'object' yang tersisa (misal: 'gender')
    original_ohe_cols = ['gender'] # Tambahkan kolom lain jika ada yang di-OHE
    for col in original_ohe_cols:
        if col in processed_df.columns and processed_df[col].dtype == 'object':
            dummies = pd.get_dummies(processed_df[col], prefix=col, drop_first=True, dtype=int)
            processed_df = pd.concat([processed_df, dummies], axis=1)
            processed_df = processed_df.drop(columns=[col])

    # --- Konversi SEMUA kolom boolean (True/False) ke int (0/1) ---
    for col in processed_df.select_dtypes(include='bool').columns:
        processed_df[col] = processed_df[col].astype(int)

    # --- Pastikan semua kolom dalam DataFrame memiliki urutan dan kelengkapan yang sama ---
    # dengan `loaded_feature_columns` (dari X_train saat training)
    final_processed_df = pd.DataFrame(columns=X_train.columns.tolist())

    for col in X_train.columns.tolist():
        if col in processed_df.columns:
            final_processed_df[col] = processed_df[col].values
        else:
            final_processed_df[col] = 0 # Tambahkan kolom yang hilang dengan 0

    # Pastikan semua kolom numerik memang numerik (setelah pemrosesan)
    for col in final_processed_df.columns:
        if col in scaler.feature_names_in_: # Cek apakah kolom ini dikenali oleh scaler
            final_processed_df[col] = pd.to_numeric(final_processed_df[col], errors='coerce')
            final_processed_df[col].fillna(0, inplace=True) # Isi NaN yang mungkin muncul dari coerce

    # --- Penskalaan Fitur Numerik (menggunakan loaded_scaler) ---
    numerical_cols_to_scale_by_scaler = scaler.feature_names_in_

    cols_to_transform = [col for col in numerical_cols_to_scale_by_scaler if col in final_processed_df.columns]

    final_processed_df[cols_to_transform] = scaler.transform(final_processed_df[cols_to_transform])

    return final_processed_df

# --- 4. Simulasikan Input Data Mentah Baru ---
# Ganti nilai-nilai ini untuk menguji skenario yang berbeda!
# Pastikan kunci (key) sesuai dengan NAMA KOLOM ASLI Anda.
# Perhatikan bahwa 'education_level' TIDAK ADA DI SINI.
new_raw_person_data = {
    'gender': 'Female', 
    'low_income': True, 
    'currently_employed': False, 
    'household_size': '5', 
    'private_vehicle': False, 
    'kks_card': True, 
    'sktm_certificate': True, 
    'dtks_registration': True, 
    'previous_social_assistance': True, 
    'stable_internet_access': False, 
    'device_access': False, # 
    'used_cek_bansos_app': False, 
    'age': '20', 
    'dependents': '5'
}

print("--- Data Mentah Baru untuk Prediksi ---")
for key, value in new_raw_person_data.items():
    print(f"{key}: {value}")

# --- 5. Preprocessing Data Baru ---
print("\n--- Memproses Data Baru... ---")
preprocessed_data_for_prediction = preprocess_single_new_data(new_raw_person_data)

print("\n--- Data Setelah Preprocessing (Siap Prediksi) ---")
print(preprocessed_data_for_prediction)
print("\nTipe Data Kolom Setelah Preprocessing:")
print(preprocessed_data_for_prediction.dtypes)


# --- 6. Melakukan Prediksi dengan Model yang Dimuat ---
print("\n--- Melakukan Prediksi Model ---")
prediction = rf_model.predict(preprocessed_data_for_prediction)
probability = rf_model.predict_proba(preprocessed_data_for_prediction)

# --- 7. Menampilkan Hasil Prediksi ---
predicted_class = int(prediction[0])
predicted_label = "Layak Menerima Bansos" if predicted_class == 1 else "Tidak Layak Menerima Bansos"
probability_of_eligibility = probability[0][1] # Probabilitas untuk Kelas 1 (True/Layak)

print("\n--- Hasil Prediksi Akhir ---")
print(f"Prediksi Kelayakan: **{predicted_label}**")
print(f"Probabilitas Kelayakan (Class 1 / Layak): **{probability_of_eligibility:.4f}**")
print(f"Probabilitas Tidak Layak (Class 0 / Tidak Layak): {probability[0][0]:.4f}")
```
