```python
import pandas as pd

# Load the CSV file (ensure correct encoding)
df = pd.read_csv('response.csv', encoding='iso-8859-1')

# Print the original column names to verify (VERY IMPORTANT)
print("Original column names:", df.columns)

# Rename columns for easier access (using the corrected names from print(df.columns))
df_new_column = df.rename(columns={
    'Timestamp': 'timestamp',
    'Usia': 'age',
    '  Jenis Kelamin  ': 'gender',
    '  Jenjang Pendidikan  ': 'education_level',
    'Apakah Anda berasal dari keluarga berpenghasilan di bawah Rp 3.500.000 per bulan?  ': 'low_income',
    'Apakah Anda bekerja saat ini?  ': 'currently_employed',
    'Jumlah anggota keluarga serumah (termasuk Anda):  ': 'household_size',
    'Jumlah tanggungan orang tua (yang belum mandiri):  ': 'dependents',
    'Apakah keluarga Anda memiliki kendaraan pribadi?  ': 'private_vehicle',
    'Apakah Anda memiliki Kartu Indonesia Pintar (KIP)?  ': 'kip_card',
    'Apakah keluarga Anda memiliki Kartu Keluarga Sejahtera (KKS)?  ': 'kks_card',
    'Apakah Anda memiliki Surat Keterangan Tidak Mampu (SKTM)?  ': 'sktm_certificate',
    'Apakah Anda atau keluarga Anda terdaftar dalam Data Terpadu Kesejahteraan Sosial (DTKS)?  ': 'dtks_registration',
    'Apakah Anda pernah menerima bantuan sosial dari pemerintah sebelumnya?  ': 'previous_social_assistance',
    'Jika ya, jenis bantuan apa yang pernah Anda terima? (bisa pilih lebih dari satu)': 'previous_assistance_types',
    'Apakah Anda memiliki akses internet yang stabil?  ': 'stable_internet_access',
    'Apakah Anda memiliki perangkat untuk mengakses aplikasi/situs bansos (HP/Laptop)?  ': 'device_access',
    'Apakah Anda sudah pernah menggunakan aplikasi âCek Bansosâ Kemensos?  ': 'used_cek_bansos_app',
    'Menurut Anda sendiri, apakah Anda merasa layak untuk menerima bantuan sosial dari pemerintah?  ': 'feel_eligible_for_assistance'
}) 

# Print the new column names to confirm the change
print("New column names:", df.columns)
```

```python
import pandas as pd

# Mapping of numerical ages to ranges (adjust these ranges as needed)
age_mapping = {
    18: '18 - 20',
    19: '18 - 20',
    20: '18 - 20',
    21: '21 - 22',
    22: '21 - 22',
    23: '23 - 25',
    24: '23 - 25',
    25: '23 - 25',
    26: '26 - 30',
    27: '26 - 30',
    28: '26 - 30',
    29: '26 - 30',
    30: '26 - 30',
}

# Function to convert age to range
def map_age_to_range(age):
    try:
        age = int(age)  # Try converting to integer
        return age_mapping.get(age)  # Return range from mapping, or None if not found
    except ValueError:
        return age  # If not an integer, return the original value

# Apply the mapping to the 'age' column
df_new_column['age'] = df_new_column['age'].apply(map_age_to_range)

# Print unique values in 'age' to verify
print(df_new_column['age'].unique())
df_new_column['age']
```

```python


# Fill NaN values in 'dependents' column with 0
df_new_column['dependents'] = df_new_column['dependents'].fillna(0)

# Mapping of numerical dependents to ranges (adjust these ranges as needed)
dependents_mapping = {
    1: '1-2',
    2: '1-2',
'satu': '1-2',
    3: '3-4',
    4: '3-4',
    5: '5-6',
    6: '5-6',
    0: '0',  # Added mapping for 0 dependents
}

# Function to convert dependents to range
def map_dependents_to_range(dependents):
    try:
        dependents = int(dependents)  # Try converting to integer
        return dependents_mapping.get(dependents)  # Return range from mapping, or None if not found
    except ValueError:
        return dependents  # If not an integer, return the original value

# Apply the mapping to the 'dependents' column
df_new_column['dependents'] = df_new_column['dependents'].apply(map_dependents_to_range)

# Print unique values in 'dependents' to verify
print(df_new_column['dependents'].unique())
```
